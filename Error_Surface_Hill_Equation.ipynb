```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Fixed dose and corresponding true response
dose = 2.0
y_true = 0.6  # "Observed" response at this dose

# Define the grid of ED50 and Hill coefficient n values
ED50_vals = np.linspace(0.1, 5, 100)
n_vals = np.linspace(0.1, 3, 100)
ED50_grid, n_grid = np.meshgrid(ED50_vals, n_vals)

# Compute predicted response using the Hill equation
predicted = 1 / (1 + (ED50_grid / dose)**n_grid)

# Compute squared error between predicted and true response
error = (predicted - y_true)**2

# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surface = ax.plot_surface(ED50_grid, n_grid, error, cmap='viridis', edgecolor='none')

# Axis labels
ax.set_title('Error Surface for Hill Equation (Varying ED50 and n)', fontsize=14)
ax.set_xlabel('ED50')
ax.set_ylabel('Hill coefficient (n)')
ax.set_zlabel('Squared Error')

# Color bar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10, label='Error Magnitude')

plt.tight_layout()
plt.show()
```
